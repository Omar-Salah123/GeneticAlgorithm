```python
import random
import numpy as np
population_size=500
mut_rate=0.1
target="computational cognitive science"
genes= " abcdefghijklmnopqrstuvwxyz"
```

```python

def fitness(Chromosome):
        error = 0
        for i in range(len(Chromosome)):
            if Chromosome[i] != target[i]:
                error +=1       
        if error == 0:
             print("Found the target string: ", Chromosome)
             print("Population size: ", population_size)
             exit(0)
        # else:
        #       fitness = 1/error
        fitness = 1/(error + 1)
        return fitness
def initialization(population_size, genes):
        chromosomes = []
        fitness_scores = []
        for i in range(population_size):
            chromosome = ""
            for j in range(len(target)):
                chromosome += random.choice(genes)
            chromosomes.append(chromosome)
            fitness_scores.append(fitness(chromosome))
        return chromosomes, fitness_scores
def selection(chromosomes,fitness_scores):
        selected = []
        pairs = zip(chromosomes, fitness_scores)
        sorted_pairs = sorted(pairs, key=lambda x: x[1], reverse=True)
        selected = sorted_pairs[:population_size//2]
        selected = np.array(selected)
        selected = selected[:, 0]
        selected = selected.tolist()
        return selected
def crossover(chromosomes, selected):
        children = []
        while len(children) < population_size:
                child =""
                parent1 = random.choice(selected)
                parent2 = random.choice(selected)
                #print("parent1", parent1)
                #print("parent2", parent2)
                mid = len(target)//2
                child += parent1[mid:] + parent2[:mid]
                #print("child", child)
                if len(child) != len(target):
                        print("Error: child length does not match target length")
                        exit(1)
                children.append(child)
        children_fitness = []
        for i in range(len(children)):
                children_fitness.append(fitness(children[i]))
        children = zip(children, children_fitness)
        children = sorted(children, key=lambda x: x[1], reverse=True)
        children = children[:population_size//2]
        children = np.array(children)
        children = children[:, 0]
        children = children.tolist()
        return children
def mutation(children):
        for i in range(len(children)):
            child = list(children[i])
            for j in range(len(child)):
                if random.random() <= mut_rate:
                        child[j] = random.choice(genes)
            children[i] = "".join(child)
        return children
```

```python
chromosomes, fitness_scores = initialization(population_size, genes)
i = 0
while i < 3000:
    print("Generation: ", i)
    i+= 1
    selected = selection(chromosomes, fitness_scores)
    children = crossover(chromosomes, selected)
    children = mutation(children)
    chromosomes = selected[:population_size//2] + children
    fitness_scores = [fitness(chrom) for chrom in chromosomes]
```

```python
print("Final population: ", chromosomes)
print("Final fitness scores: ", fitness_scores)
print("Best fitness score: ", max(fitness_scores))
print("currrent population size: ", len(chromosomes))
best_chromosome = chromosomes[fitness_scores.index(max(fitness_scores))]
print("Best chromosome: ", best_chromosome)
```

```python
for i in range(len(target)):
    if best_chromosome[i] == target[i]:
        print("Correct gene at position ", i, ": ", best_chromosome[i])
    else:
        print("Incorrect gene at position ", i, ": ", best_chromosome[i], " should be:", target[i])
```
